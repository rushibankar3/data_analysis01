```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
plt.show
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
plt.show
```

```python
gap = px.data.gapminder()
gap.head()
```

```python

temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

```python
temp_df=gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df
```

```python
sns.relplot(kind='line',data=temp_df,x='year',y='lifeExp',hue='country',style='continent',size='continent')
plt.show()
```

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp',hue='country')
plt.show()
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='line',col='day',row='sex')
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", bins=20, color="skyblue")
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", bins=20, kde=True, color="skyblue")
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',bins=2,color='yellow')
plt.show()
```

```python
sns.displot(data=tips,kind='hist',x='day',bins=5)
plt.show
```

```python
sns.displot(data=tips, x='tip', kind='hist',hue='sex')
plt.show()
```

```python
sns.displot(data=tips,x='tip',kind='hist',hue='sex',element='step')
plt.show()
```
